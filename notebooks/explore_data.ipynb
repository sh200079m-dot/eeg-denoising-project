```python
import os
import mne
import matplotlib.pyplot as plt
import numpy as np

# Load the EEG dataset from a .set file (EEGLAB format)
file_path = "/mnt/d/EEG_Project/data/sub-001/eeg/sub-001_task-Rest_eeg.set"
raw = mne.io.read_raw_eeglab(file_path, preload=True)
# Print information about the loaded raw data (channels, sampling frequency, etc.)
print(raw.info)
# Print the shape of the data array (channels  time points)
print(f"Shape: {raw.get_data().shape}")

# Plot the first 10 channels for 5 seconds with automatic scaling
raw.plot(n_channels=10, duration=5, scalings="auto")
# Display the plot
plt.show()
```
